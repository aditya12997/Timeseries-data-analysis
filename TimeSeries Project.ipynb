## Let's plot the closing stock prices of the various companies to understand how the trend line is

#### We are using Google, Amazon, Apple, microsoft data from yahoo finance to compare the stock prices

```python
google_df = pd.read_csv("GOOG.csv")
microsoft_df = pd.read_csv("MSFT.csv")
amazon_df = pd.read_csv("AMZN.csv")
apple_df = pd.read_csv("AAPL.csv")

google_close = google_df['Close']
microsoft_close = microsoft_df['Close']
amazon_close = amazon_df['Close']
apple_close = apple_df['Close']

# for sample to see how the data looks like for all companies
google_df.tail(10)
```

```python
#setting the period one year
lookback = 365

# creating the ROI for each company so we can plot them
ROI_1_google = 100 * (google_close.shift(lookback) / google_close -1)
ROI_1_microsoft = 100 * (microsoft_close.shift(lookback) / microsoft_close -1)
ROI_1_amazon = 100 * (amazon_close.shift(lookback) / amazon_close -1)
ROI_1_apple = 100 * (apple_close.shift(lookback) / apple_close -1)


#Finally plotting them to see the performance of these companies
fig, ax = plt.subplots(figsize = (15,7))
ax.plot(google_close.index, ROI_1_google, label = "GOOGLE")
ax.plot(microsoft_close.index, ROI_1_microsoft, label = "MICROSOFT")
ax.plot(amazon_close.index, ROI_1_amazon, label = "AMAZON")
ax.plot(apple_close.index, ROI_1_apple, label = "APPLE")
plt.title("ROI of different companies in the last 1 year",fontsize=20)
plt.xlabel('% of ROI',fontsize=15)
plt.ylabel("Time",fontsize=15)
plt.legend(fontsize=10)
plt.show()
```

```python
#setting the period one year
# we're using the last 100 days here
lookback = 100

# creating the ROI for each company so we can plot them
ROI_1_google = 100 * (google_close.shift(lookback) / google_close -1)
ROI_1_microsoft = 100 * (microsoft_close.shift(lookback) / microsoft_close -1)
ROI_1_amazon = 100 * (amazon_close.shift(lookback) / amazon_close -1)
ROI_1_apple = 100 * (apple_close.shift(lookback) / apple_close -1)


#Finally plotting them to see the performance of these companies
fig, ax = plt.subplots(figsize = (15,7))
ax.plot(google_close.index, ROI_1_google, label = "GOOGLE")
ax.plot(microsoft_close.index, ROI_1_microsoft, label = "MICROSOFT")
ax.plot(amazon_close.index, ROI_1_amazon, label = "AMAZON")
ax.plot(apple_close.index, ROI_1_apple, label = "APPLE")
plt.title("ROI of different companies in the last 100 Days (More Granularity)",fontsize=20)
plt.xlabel('% of ROI',fontsize=15)
plt.ylabel("Time",fontsize=15)
plt.legend(fontsize=15)
plt.show()
```
