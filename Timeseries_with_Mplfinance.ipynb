### In this NoteBook we are using Pandas and MPLFinance to plot our Timeseries Data and Understand the Share Price of Amazon in a Particular period of time.

We are going to see if there was any shift in the curve BEFORE and AFTER COVID-19. 

Did the share price really fall?

Was there an Impact as such? Let's see ðŸ‘€ 

```python
import pandas as pd
import mplfinance as mpl

# if you can't access the Mpl library use the command in your notebook as "pip install --upgrade mplfinance"
```

```python
df = pd.read_csv("AMZN.csv")
df.head()
```

### Let's look at our data and it's DataTypes

```python
df.info()
```

As we see that the data column is not indexed and it's in the wrong datatype. 

Hence we need to convert it into a datetime series with the help of pandas.

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df.set_index("Date")
```

```python
df.head()
```

### Plotting to data to see how the visuals look like:

```python
mpl.plot(df)
```

###### The share price for a period of time, as is the definition of Timeseries

##### The following way shows how can we plot the price and volume of the shares below for clear understanding of it's performance in a period of time.

```python
mpl.plot(df, type ='candle', volume =True)
```

### Below we are going to see if there was a unusual dip on the curve or not due to COVID-19


Yes, there is a dip in the shareprice when COVID-19 initially affected the whole world

```python
mpl.plot(df['2020-01':'2020-04'],type ='line',
         figratio =(20,10), volume =True,
         title = "Amazon Share Price before and After COVID-19")
```

### Setting the Moving Average to look for any short-term fluctuations on the price of a stock over a specified time frame are mitigated.

```python
mpl.plot(df['2020-01':'2020-04'],type ='candle', mav = (20),
         figratio =(20,10), volume =True,
         title = "Amazon Share Price before and After COVID-19", style ="yahoo")
```
